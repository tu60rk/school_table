# Модули генетического алгоритма 

### Создание индивидума 

```python
def create_individum() -> list:
    """
    Создаем индивидума. 1 индивидум - 1 школьное расписание. Расписание предствляет собой матрицу 6 на 8 
    (6 дней в неделю на обучение и 8 уроков каждый день).
    Заполняем рандомными курсами индивидума.
    """
    individum = []
    for lesson in CLASS_TABLE:
        for class_room in CLASSROOMS:
            individum.append(np.random.randint(0, len(COURSES)))
    return individum 
```

### Создание популяции 

```python
def create_population(SIZE_POPULATION : int) -> list:
    """
    Создаем популяцию. Заполняем ее индивидумами до определенного значения.
    """
    return [create_individum() for _ in range(SIZE_POPULATION)]
```

### Функция минимизации 

```python
def cost(individum : list) -> int:
    """
    Определим функцию стоимости индивидума, которую в будущем будем минимизировать.
    В функцию стоимости заходят жесткие ограничения
    """
    value = hardConstraintsCountLessons(individum) + \
            hardConstraintsDifferentClasses(individum) + \
            hardConstraintsOneTeacher(individum)
    return value
```

### Жесткие ограничения функции минимизации 

```python
def hardConstraintsCountLessons(individum : list) -> int:

    """
    Функция дает штраф, если мы поставили больше или меньше занятий по курсу.
    """
    
    count_violation = []
    for curs in COURSES.keys():
        countCourseNow = individum.count(curs)
        countCourseMust = COURSES[curs].count_lessons
        
        count_violation.append(abs(countCourseMust - countCourseNow))
    return sum(count_violation)
```

```python
def hardConstraintsDifferentClasses(individum : list) -> int:
    
    """
    Функция дает штраф, если один и тот же курс преподается в один и тоже урок
    """
    
    count_violation = 0
    length_step = len(CLASSROOMS)
    for i in range(0, len(individum), length_step):
        if len(set(individum[i: i + length_step])) == len(individum):
            continue
        else:
            count_violation += 1
    return count_violation
```

```python
def hardConstraintsOneTeacher(individum : list) -> int:
    
    """
    Функция дает штраф, если один и тот же учитель занят в один урок на разных курсах
    """
    
    count_violation = 0
    length_step = len(CLASSROOMS)
    for i in range(0, len(individum), length_step):
        teachers_lessons = [COURSES[curs].TeacherFIO for curs in individum[i: i + length_step]]
        if len(set(teachers_lessons)) != len(individum):
            count_violation += 1
            
    return count_violation
```

### Функция отбора индивидумов 

```python
def select(population : list, COSTS : list, size_tournament : int) -> list:
    
    """
    Функция выбирает наилучшего индивидума путем турнирного метода.
    Возвращает список наилучших индивидумов
    """
    
    k = len(population) / size_tournament
    best = []
    for _ in range(int(k)):        
        keys_lst = [i for i in range(len(COSTS))]
        idx_list = [random.choice(keys_lst) for _ in range(size_tournament)]       
        best.append(population[sorted([ [idx, COSTS[idx]] for idx in idx_list ], key = lambda x: x[1])[0][0]])
        
    best_costs = sorted([[ind, value] for ind, value in zip(best, [cost(ind) for ind in best])], key = lambda x: x[1])
    return [ind[0] for ind in best_costs]
```

### Скрещивание индивидумов 

```python
def mate(population : list, best_population : list, 
         P_CROSSOVER : float, SIZE_POPULATION : int, P_CROSSOVER_FIRST_PARENT : float) -> list:
    
    """
    Функция, которая скрещивает популяцию.
    Возвращаем уже скрещенную популяцию.
    """
    
    new_population = []
    while len(new_population) < SIZE_POPULATION:
        random.shuffle(population)
        for parent1, parent2 in zip(best_population, population):
            if random.random() < P_CROSSOVER:
#  Равномерное скрещивание
#                 new_population.append([np.random.choice([par1, par2], 
#                                                         p = [P_CROSSOVER_FIRST_PARENT, 1 - P_CROSSOVER_FIRST_PARENT])\
#                                        for par1, par2 in zip(parent1, parent2)])
                new_population.append([random.choice([par1, par2]) for par1, par2 in zip(parent1, parent2)])
                new_population.append([par2 if add_par == par1 else par1 for add_par, par1, par2 in zip(new_population[-1], parent1, parent2)])
#  Одноточечное скрещивание
#                 point = random.randint(0, len(parent1))
#                 new_population.append(parent1[:point] + parent2[point:])
#                 new_population.append(parent2[:point] + parent1[point:])
#  Двуточечное скрещивание
#                 point_one = random.randint(0, len(parent1))
#                 point_two = random.randint(point_one, len(parent1))
#                 new_population.append(parent1[point_one:point_two] + parent2[:point_one] + parent2[point_two:])
#                 new_population.append(parent2[point_one:point_two] + parent1[:point_one] + parent1[point_two:])
            else:
                new_population.append(parent1)
                new_population.append(parent2)
        
    return new_population
```

### Мутация индивидумов 

```python
def mutate(population : int, P_MUTATE : float) -> list:
    
    """
    Мутация внутри новой популяции.
    Возвращает популяцию.
    """
    
    new_population = []
    for mutant in population:
        if random.random() < P_MUTATE:
            idxChange = random.randint(0, len(mutant))
            part_mutant = mutant[idxChange:]
            random.shuffle(part_mutant)
            mutant = mutant[:idxChange] + part_mutant
        new_population.append(mutant)
    return new_population 
```

# Генетический алгоритм 

```python
def main() -> None:
    """
    Функция запускает генетический алгоритм.
    Возвращает график изменения приспособленности популяций.
    """
    
    # ***********************
    # Обязательные параметры, которые можно и нужно менять !
    
    np.random.seed(42) # чтобы рандом начинался не с нуля
    random.seed(42)    # чтобы рандом не начинался с нуля
     
    SIZE_POPULATION = 50   # Размер популяции
    MAX_GENERATIONS = 4000 # Шаги эволюций
    P_CROSSOVER_FIRST_PARENT = 0.7  # Вероятность наследования от первого родителя
    
    P_CROSSOVER = 0.9 # Вероятность скрещивания
    P_MUTATION = 0.5  # Вероятность мутации
    
    # Обязательные параметры конец
    # ***********************
    
    POPULATION = create_population(SIZE_POPULATION)
    
    min_costs = []
    avg_costs = []
    
    for i in tqdm(range(MAX_GENERATIONS)):
        
        COSTS = [cost(ind) for ind in POPULATION]
        NEW_POPULATION = select(POPULATION, COSTS, 2)
        POPULATION = mate(POPULATION, NEW_POPULATION, P_CROSSOVER, SIZE_POPULATION, P_CROSSOVER_FIRST_PARENT)
        POPULATION = mutate(POPULATION[:SIZE_POPULATION], P_MUTATION)
        
        min_costs.append(min(COSTS))
        avg_costs.append(np.mean(COSTS))
        
    
    plt.plot(min_costs, color='red')
    plt.plot(avg_costs, color='green')
    plt.xlabel('Поколение')
    plt.ylabel('Макс/средняя приспособленность')
    plt.title('Зависимость максимальной и средней приспособленности от поколения') 
    plt.show()     
```
